## <div style="color:#fbb144;"> 1. Background </div>


Biosignals have been extensively used in the healthcare and medical domains for more than 100 years, with the best known examples being perhaps Electrocardiography (ECG) and Electroencephalography (EEG). In this lesson we will perform a set of experiments using a real-world biomedical data acquisition system, understand the basic architecture of such systems, review typical setup procedures, and observe multiple signal sources in real-time.

## <div style="color:#fbb144;"> 2. Objectives</div>
 * Understand the basic blocks of a data acquisition system
 * Get acquainted with biosignal acquisition procedures
 * Handling recorded data on a computing environment
