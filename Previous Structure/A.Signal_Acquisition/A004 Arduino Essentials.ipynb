## <div style="color:#fbb144;"> 1. Background </div>


The Arduino is an open source hardware and software platform, designed to empower virtually anyone with the basic tools to easily create microcontroller-
based devices. In this lesson we will dive into the world of embedded systems
programming, setup a development environment, and get acquainted with the
Arduino platform and programming paradigm. Note that there are dozens of
variants of this platform and we will use only one of them. The addressed topics
and base principles are common to all, but keep in mind to throroughly review
the specifications of the devices you use to make the most of their capabilities.

<img src="https://github.com/PIA-Group/ScientIST-notebooks/blob/master/_Resources/Images/A.Signal_Acquisition_IMG/a004/ArduinoUNO.png?raw=true" width=400/>

## <div style="color:#fbb144;"> 2. Objectives</div>
* Understand the Arduino platform and computational paradigm
* Setup the development tools in your computer
* Write, compile, program and execute code on the Arduino

  <div style="background:#ffdab0;font-size:12px"> 
NEVER:
    <li> connect the pins labelled as POWER directly between them (e.g. the 3.3V pin to the 5V pin).
 <li> connect the voltage pins to another (external) voltage source, regardless of its type. 
<li>connect the voltage pins to the GND. 
<li>leave unconnected (aka “live”) wires floating around your board as they can inadvertently create short circuits between components.
<li> bend, twist, cut, score, apply weight or otherwise perform any mechanical abuse to your board; it is fragile! <br>
ALWAYS:
    <li> have your board disconnect from the computer while assembling your circuits! 
</div>

## <div style="color:#fbb144;">  4. Embedded Hello World </div>


One of the simplest programs that can be created using an embedded system is
blinking an LED; you can see it almost as the hardware equivalent to the “Hello
World” program normally used in most programming languages to illustrate
the basic syntax. 
<br>
For this experiment you will use one of the LEDs integrated
on your board, which is connected to the GPIO 13 (constant
LED BUILTIN), as a way of becoming familiar with the structure of an Arduino
sketch.

    Location of the LED on your board:
<img src="https://github.com/PIA-Group/ScientIST-notebooks/blob/master/_Resources/Images/A.Signal_Acquisition_IMG/a004/ArduinoLED.png?raw=true" width="275"> 

    How to access the Blink example:
<img src="https://github.com/PIA-Group/ScientIST-notebooks/blob/master/_Resources/Images/A.Signal_Acquisition_IMG/a004/Blink.png?raw=true" width="275"> 

    1. Load the Blink example sketch and review the code
    2. Compile and upload the sketch to your Arduino board
    3. Upon successful completion of the process, the LED associated with GPIO 13 should start flashing at 0.5Hz

## <div style="color:#fbb144;">  5. Multichannel Digital Control</div>


Now that we’ve seen how to do a simple control task, let’s take it to the next level
and control several channels. In this experiment, you will use three independent
LEDs to create the base for a three level Visual Analog Scale (VAS) display.
    1. Take the breadboard, the 330Ω resistors, the jumper wires and the red, yellow and green LEDs, and connect them to your Arduino

        Controlling multiple LEDs with the Arduino:
<img src="https://github.com/PIA-Group/ScientIST-notebooks/blob/master/_Resources/Images/A.Signal_Acquisition_IMG/a004/MultiBlink.png?raw=true" width="400"> 

    2. Adapt the Blink example sketch to control the pins to which each LED is connected, lighting them in sequence; use an 
    array and a repetition structure to avoid code duplication
    3. Compile and upload the sketch to your Arduino board
    4. Upon successful completion of the process, each LED will continuously turn on in sequence

<div style="background:#fbb144;font-family:'arial', monospace; text-align: center; padding: 10px 0; border-radius:10px; width:70%; margin:auto " >
  <span style="font-size:20px;position:relative;color:white; ">  Warning! </span> <br>
  <div style="background:#ffd08a;font-size:12px"> 
   The LEDs have polarity and need to be connected in a specific way in
order to work properly. The long leg is the power (+) line and the short
leg is the ground (-) line.
</div>

## <div style="color:#fbb144;">  6. Reading Digital Inputs</div>


Until now we focused only on controlling outputs, but sensing is equally (if
not more) important. In this experiment we’ll be looking into one of the most
common and simple input methods - the push button.
    1. Take the breadboard, the 10kΩ resistor, the jumper wires and the push button, and connect them to your Arduino
    2. Load the Button example sketch and review the code
    3. Compile and upload the sketch to your Arduino board
    4. Upon successful completion of the process, the LED associated with GPIO 13 should turn off whenever you push the
    button
