## <div style="color:#fbb144;"> 1. Background </div>

Google Science Journal (GSJ) is an App designed for students. It is able to transforma a phone, tablet, and Chromebook into a science notebook that encourages students to explore their world. 

GSJ allows students to measure their experiment results, take quality photos and make annotations throughout their experiments. Device sensors can be used for data acquisition, e.g. to measure light, sound, or movements.

## <div style="color:#fbb144;">  1. Getting Started </div>

1. Locate and launch the GSJ app; you can sign in with your Google account or proceed without signing in. 

2. On the GSJ app home screen (next figure, left) a library is shown with the history of experiments previously created, ordered by date of last edit; before any further experiments are created, only the “Welcome to Science Journal” entry appears.
3. Create a new experiment using the button (+).
4. A journal annotation screen appears (next figure, right)!

## <div style="color:#fbb144;">  2. Creating a Basic Journal </div>

1. Access the journal settings screen using the "pencil" button.
2. Define a title for your experiment.
3. Assign a cover photo to your experiment.
4. Save the new settings using the "check" button.
5. In the journal you can create observations using multiple formats; these will appear in the top of the screen.
6. Text observations can be written in the text pane that appears when the "comment" button is pressed and submitted to the journal pressing the "right arrow" button.
7. Photos and screenshots can also be added to the journal by pressing the "photo" button (to take a photo) or the "image" button (to choose from the gallery); in the later case the image is submitted to the journal pressing the "right arrow" button.

## <div style="color:#fbb144;">  4. Sharing your Data </div>

1. You can download the complete journal or share it (if you are using the GSJ with a Google account) using the options available through the menu on the main screen
2. To export recorded time series tap the graph with the recording you would like to export to access the reviewing screen (next figure, left), and follow the next steps
3. Use the buttons a n d to browse through the channels (next figure, right)!
4. You can slide through the timeline and add annotations using the “ADD NOTE TO TIMELINE” option (next figure, right)!
5. Data can be downloaded as Comma-Separated Values (CSV) through the corresponding option on the menu
6. If you are using the GSJ with a Google account, it will also be possible to share the recording directly to the Google Drive, by e-mail, and other methods
7. When exporting your data, enabling the “Relative time” option is recommended so that the timeline is zero based, otherwise time will be exported as Unix Timestamps

## <div style="color:#fbb144;">  5. Your First Experiment </div>

1. Create and configure the settings for a new journal to document a motion data acquisition experiment, complete with an appropriate title and cover photo
2. Complement the following steps with suitable textual and pictorial annotations that can thoroughly describe the materials, experimental methods, and data collected
3. Add sensor observation cards to your journal for recording Accelerometer X, Accelerometer Y, Accelerometer Z and sound data; the later should be used for event “annotation”
4. Hold the phone in your hand, start recording the data, perform 10 repetitions of right arm elevation at 90o in the coronal plane, and stop the recording
5. Repeat the previous step for 180o elevations
6. Repeat step 4. for the sagittal plane
7. Repeat step 4. for the sagittal plane and 180o elevations
8. Share your science journal, a PDF report (only available on iOS), and data in CSV format with your lecturer
