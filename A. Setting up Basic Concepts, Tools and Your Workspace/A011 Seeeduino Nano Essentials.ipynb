## <div style="color:#fbb144;"> 1. Background </div>


The Seeeduino Nano is an Arduino-derived open source hardware and software platform, designed to empower virtually anyone with the basic tools to easily create microcontroller-based devices. In this lesson we will dive into the world of embedded systems programming, setup a development environment, and get acquainted with the Arduino platform and programming paradigm. Note that there are dozens of variants of this platform and we will use only one of them. The addressed topics and base principles are common to all, but keep in mind to throroughly review the specifications of the devices you use to make the most of their capabilities.

<img src="https://github.com/PIA-Group/ScientIST-notebooks/blob/master/_Resources/Images/A.Signal_Acquisition_IMG/a011/seeeduino.png?raw=true" width=400/>

## <div style="color:#fbb144;"> 2. Objectives</div>
* Understand the Arduino platform and computational paradigm
* Setup the development tools in your computer
* Write, compile, program and execute code on the Seeeduino Nano

NEVER:
    <li> Connect the pins labelled as POWER directly between them (e.g. the 3.3V pin to the 5V pin).
 <li> Connect the voltage pins to another (external) voltage source, regardless of its type. 
<li> Connect the voltage pins to the GND. 
<li> Leave unconnected (aka “live”) wires floating around your board as they can inadvertently create short circuits between components.
<li> Bend, twist, cut, score, apply weight or otherwise perform any mechanical abuse to your board; it is fragile! <br>
ALWAYS:
    <li> Have your board disconnect from the computer while assembling your circuits!

## <div style="color:#fbb144;">  4. Embedded Hello World </div>


One of the simplest programs that can be created using an embedded system is
blinking an LED; you can see it almost as the hardware equivalent to the “Hello
World” program normally used in most programming languages to illustrate
the basic syntax. 
<br>
For this experiment you will use one of the LEDs integrated
on your board, which is connected to the GPIO 13 (constant
LED_BUILTIN), as a way of becoming familiar with the structure of a Seeeduino Nano 
sketch.

    Location of the LED on your board:
<img src="https://github.com/PIA-Group/ScientIST-notebooks/blob/master/_Resources/Images/A.Signal_Acquisition_IMG/a011/seeeduinoLED.png?raw=true" width="400"> 

    How to access the Blink example:
<img src="https://github.com/PIA-Group/ScientIST-notebooks/blob/master/_Resources/Images/A.Signal_Acquisition_IMG/a004/Blink.png?raw=true" width="275"> 

    1. Load the Blink example sketch and review the code
    2. Compile and upload the sketch to your Seeeduino Nano board
    3. Upon successful completion of the process, the LED associated with GPIO 13 should start flashing at 0.5Hz

1. Write an alternative formulation of the 01.BASICS/BLINK sketch that uses a single call to the digitalWrite and delay functions.

2. Explain what are the drawbacks of using the delay function and write an alternative formulations of the 01.BASICS/BLINK sketch that produces the same result without using the delay function.
